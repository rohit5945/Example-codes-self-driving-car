```python
import numpy as np
import cv2
import matplotlib.image as img
import matplotlib.pyplot as plt
%matplotlib qt
```

```python
img=img.imread('traffic.jpg')
plt.imshow(img)
```

```python
%matplotlib inline

plt.imshow(img)
plt.plot(249,83,'.')
plt.plot(250,141,'.')
plt.plot(96,136,'.')
plt.plot(97,72,'.')
```

```python
#define perspective transform
def warp(img):
    
    img_size=(img.shape[1],img.shape[0])
    
    src=np.float32(
        [[249,83],
         [250,141],
         [96,136],
         [97,72]
        ])
    #desired coordinates
    dsrd=np.float32(
        [[200,83],
         [200,141],
         [80,136],
         [80,72]
        ])
    M=cv2.getPerspectiveTransform(src,dsrd)
    Minv=cv2.getPerspectiveTransform(dsrd,src)
    warped=cv2.warpPerspective(img,M,img_size,flags=cv2.INTER_LINEAR)
    return warped
```

```python
%matplotlib inline
warped_im=warp(img)
f,(ax1,ax2)=plt.subplots(1,2,figsize=(20,10))
ax1.imshow(img)
ax2.imshow(warped_im)
```
